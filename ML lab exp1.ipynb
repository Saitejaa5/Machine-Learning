```python
import pandas as pd


data = {
    "StudentID": [1, 2, 3, 4, 5, 6, 7, 8],
    "ExamScore": [85, 78, None, 92, 88, None, 75, 80],
    "GradeLevel": ["A", "B", "B", "A", "A", "C", "C", "B"]
}
df = pd.DataFrame(data)

print("Original Dataset:")
print(df)


df['ExamScore_Mean'] = df['ExamScore'].fillna(df['ExamScore'].mean())
df['ExamScore_Median'] = df['ExamScore'].fillna(df['ExamScore'].median())
exam_score_mode = df['ExamScore'].mode()[0]
df['ExamScore_Mode'] = df['ExamScore'].fillna(exam_score_mode)

print("\nDataset after cleaning 'ExamScore'with mean, median, and mode:")
print(df[['StudentID', 'ExamScore', 'ExamScore_Mean', 'ExamScore_Median', 'ExamScore_Mode', 'GradeLevel']])
```
