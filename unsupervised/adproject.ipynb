This code checks for missing values
* fills numerical columns with their median
* fills categorical columns with their mode
* Then it verifies that all missing values are handled.

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler


print("Missing values per column:\n", df.isnull().sum())

numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[numerical_cols] = df[numerical_cols].fillna(df[numerical_cols].median())


categorical_cols = df.select_dtypes(include=['object']).columns
df[categorical_cols] = df[categorical_cols].fillna(df[categorical_cols].mode().iloc[0])

print("\nMissing values after handling:\n", df.isnull().sum())
```

## This part encodes categorical variables into numerical values using LabelEncoder. 
* It iterates through all categorical columns
* converts each unique category into a numeric label.
* Finally, it prints a sample of the transformed dataset.

```python
label_encoder = LabelEncoder()

for col in categorical_cols:
    df[col] = label_encoder.fit_transform(df[col])

print("\nSample data after encoding categorical variables:")
df.head()
```

## This code normalizes numerical features (excluding the 'ID' column) using StandardScaler
* It scales the data to have a mean of 0 and a standard deviation of 1.
* It then prints a sample of the normalized dataset.

```python
features_to_normalize = [col for col in numerical_cols if col != 'ID']

scaler = StandardScaler()
df[features_to_normalize] = scaler.fit_transform(df[features_to_normalize])

print("\nSample data after normalization:")
df.head()
```

## This code calculates two new features:

* Total_Spending: The sum of spending across different product categories (MntWines, MntFruits, MntMeatProducts, etc.).
* Total_Children: The total number of children in a household by adding Kidhome (young children) and Teenhome (teenagers).
  
#### These features help in analyzing customer spending behavior and household composition.

```python
import datetime


spending_cols = ['MntWines', 'MntFruits', 'MntMeatProducts', 
                 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
df['Total_Spending'] = df[spending_cols].sum(axis=1)

df['Total_Children'] = df['Kidhome'] + df['Teenhome']
```

## Principal Component Analysis (PCA) is performed on the dataset to analyze the explained variance of different components.

* Fit PCA: PCA().fit(df) computes principal components without reducing dimensions yet.
* Plot Cumulative Explained Variance:
  
> X-axis: Number of principal components.

> Y-axis: Cumulative explained variance (how much information is retained as components increase).
* Helps determine the optimal number of components for dimensionality reduction.
* This visualization guides in selecting the right number of components to balance information retention and dimensionality reduction.

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt


pca_full = PCA().fit(df)
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(df.columns)+1), pca_full.explained_variance_ratio_.cumsum(), marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance vs. Number of Components')
plt.grid(True)
plt.show()
produces a line plot
```

## This code applies the Elbow Method to find the optimal number of clusters for K-Means clustering. 
* It plots k (number of clusters) vs. inertia (WCSS). 
* The elbow point in the graph indicates the best k.
* where adding more clusters no longer significantly reduces inertia.

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import pandas as pd
# range for k 
cluster_range = range(1, 11)
inertia_values = []

# K-Means clustering 
for k in cluster_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10) # groups similar data point together into k cluster
    kmeans.fit(df)
    inertia_values.append(kmeans.inertia_)


plt.figure(figsize=(10, 6))
plt.plot(cluster_range, inertia_values, marker='o', linestyle='--', color='green', linewidth=2, markersize=6)
plt.xlabel('Number of Clusters (k)', fontsize=12)
plt.ylabel('Inertia (WCSS)', fontsize=12)
plt.title('Elbow Method for Optimal k (K-Means)', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Elbow Method graph
```

#### output: 
* An Elbow Curve, which is a line plot 
* shows the relationship between the number of clusters (k) and inertia (WCSS - Within-Cluster Sum of Squares).

## This code applies Hierarchical Clustering to a random sample of 500 rows using Ward’s method 
* It visualizes the clusters with a dendrogram. 
* The dendrogram shows how data points merge into clusters
* It helping determine the optimal number of clusters based on distance.

```python
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

# in hierachical clustering we dont need to specify no. of clusters beforehand

df_sample = df.sample(n=500, random_state=42)

linked = linkage(df_sample, method='ward')

plt.figure(figsize=(12, 6))
dendrogram(linked, truncate_mode='lastp', p=12, leaf_rotation=45, leaf_font_size=12, show_contracted=True)
plt.title('Dendrogram (Hierarchical Clustering on Sample)', fontsize=14)
plt.xlabel('Cluster Size', fontsize=12)
plt.ylabel('Distance', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

## This code applies DBSCAN (Density-Based Spatial Clustering) and helps determine the optimal eps value using the k-distance graph.
* DBSCAN clustering: It groups similar data points while detecting outliers.
* It finds the 4th nearest neighbor distance for each data point using NearestNeighbors.
* Plots the k-distance graph to help determine the eps value (elbow point).
* Applies DBSCAN clustering with eps=0.5 and min_samples=5.
* Prints the cluster distribution, showing how many points belong to each cluster, including noise (-1).
* This approach helps identify dense clusters while detecting outliers.

```python
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
import numpy as np

# it does not require specifying the number of clusters (K) and can detect outliers
# compute k-Nearest Neighbors Distances
neighbors = NearestNeighbors(n_neighbors=5)
neighbors_fit = neighbors.fit(df)
distances, indices = neighbors_fit.kneighbors(df)
sorted_distances = np.sort(distances[:, 4])  # 4th neighbor distance

#Plot the k-Distance Graph
plt.figure(figsize=(10, 6))
plt.plot(sorted_distances, marker='o', linestyle='--', color='red', markersize=6)
plt.xlabel('Data Points sorted by distance', fontsize=12)
plt.ylabel('4th Nearest Neighbor Distance', fontsize=12)
plt.title('DBSCAN: k-Distance Graph for Optimal eps', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Applying DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5) # eps is distance bw neighbours and min_samples are total no. of neighbours
dbscan_labels = dbscan.fit_predict(df)

# analyzes and prints the distribution of clusters formed by DBSCAN.
unique_labels, counts = np.unique(dbscan_labels, return_counts=True) 
print("DBSCAN Cluster Distribution:")
for label, count in zip(unique_labels, counts):
    print(f"Cluster {label}: {count} points")
```
