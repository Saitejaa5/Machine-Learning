```python
kmeans = KMeans(n_clusters = 2, random_state = 42)
```

```python
kmeans.fit(val)
```

```python
label = kmeans.labels_
label
```

```python
center = kmeans.cluster_centers_
center
```

```python
val.head()
```

```python
labels = kmeans.fit_predict(val)
val['Cluster'] = labels
```

```python
val['Cluster'].unique()
```

```python
val['Cluster'].value_counts()
```

```python
plt.scatter(val.iloc[:, 0], val.iloc[:, 1], c=labels, cmap='viridis', alpha=0.7)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='x', s=200, label='Centroids')
plt.legend()
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)  
val_pca = pca.fit_transform(val.drop(columns=['Cluster']))  

plt.scatter(val_pca[:, 0], val_pca[:, 1], c=labels, cmap='viridis', alpha=0.7)
plt.scatter(pca.transform(kmeans.cluster_centers_)[:, 0], pca.transform(kmeans.cluster_centers_)[:, 1], c='red', marker='x', s=200, label='Centroids')
plt.legend()
plt.show()
```

```python
print("Inertia:", kmeans.inertia_)
```
