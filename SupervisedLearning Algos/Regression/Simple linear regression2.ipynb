- Linear dependence between the x and y variables 
- the residual errors shall have constant variance plot residual vs homoscalesticty
- residuals shall be normally distributed

```python
#build regression model

model = smf.ols("sunday ~ daily", data=data1).fit()

# Display the model summary
print(model.summary())           
```

```python
x = data1["daily"].values
y = data1["sunday"].values
plt.scatter(x , y , color = "m", marker = "o",s = 30)
b0 = 12.84
b1 = 1.33
y_hat = b0 + b1*x

plt.plot(x,y_hat, color = "g")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
data1.corr(numeric_only=True)
```

Observations

the relationship between x(daily) and y(sunday) is seen to be linear as seen from scatter plot the correlation is strong and positive with pearsons correlation

#from the model summary we get a predicted equation like
# y_hat = 13.8356+1.3397 x
# beta_0 = 13.8356, beta_1 = 1.3397
# the probability(p-value) for intercept (beta_0) is 0.707 >      0.05
# therefore the intercept coefficient may not be that much significant in prediction.
# however the p-value for "daily" (beta_1) is 0.00 < 0.05
# therefore the beta_1 coefficient is highly significant and is contributint to prediction.

```python
import statsmodels.formula.api as smf
model1 = smf.ols("sunday~daily",data = data1).fit()



model1.summary()
```

```python
model1.params
```

```python
(model1.rsquared,model1.rsquared_adj)
```

```python
newdata=pd.Series([200,300,1500])
data_pred=pd.DataFrame(newdata,columns=['daily'])
data_pred
```

```python
model1.predict(data_pred)
```

```python
#predict on given all training data

pred = model1.predict(data1["daily"])
pred
```

```python
#add predicted values as column
data1["Y_hat"] = pred
data1
```

```python
data1["residuals"] = data1["sunday"]-data1["Y_hat"]
data1
```

```python
mse = np.mean((data1["daily"]-data1["Y_hat"])**2)
rmse = np.sqrt(mse)
print("MSE: ",mse)
print("RMSE: ",rmse)
```

```python
plt.scatter(data1["Y_hat"], data1["residuals"])
```

***Observations***
- there appears to be no trend and the residuals are randomly palces around the zero
- hence the assumptions of homoscedasticity is satisfied [constant variance in residual
